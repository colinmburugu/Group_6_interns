```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
%matplotlib inline
```

### Reading Data

```python
 
df = pd.read_stata("data/5pct_sample_2019_10.dta")
```

```python
db_date = pd.read_csv("data/db_date.csv")
```

```python
new_df = df.merge(db_date, on = 'date')
```

```python
new_df.head()
```

```python
df1 = np.where(new_df['item'] == 250)
```

```python
df1 = new_df.loc[df1]
```

```python
ts = df1[['date2','price']]
ts
```

```python
ts.rename(columns={'date2':'date'},inplace = True)
```

```python
ts['date'] = pd.to_datetime(ts["date"],format="%d-%m-%Y")
```

#### Data Cleaning and Pre-processing

```python
#setting the date column as our new index
ts.set_index("date",inplace = True)
```

```python
ts
```

```python
ts.index.duplicated()
```

```python
ts.info()
```

```python
ts = ts['price'].resample('MS').sum()
```

```python
plt.figure(figsize = (20,7))
ts.plot()
```

#### Stationarity check

```python
import statsmodels.api as sm
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
```

```python
# Multiplicative Decomposition 
multiplicative_decomposition = seasonal_decompose(ts, model='multiplicative')

# Additive Decomposition
additive_decomposition = seasonal_decompose(ts, model='additive')

# Plot
plt.rcParams.update({'figure.figsize': (16,12)})
multiplicative_decomposition.plot().suptitle('Multiplicative Decomposition', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

additive_decomposition.plot().suptitle('Additive Decomposition', fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.show()
```

```python
def adf_test(series):
    result=adfuller(series)
    print('ADF Statistics: {}'.format(result[0]))
    print('p- value: {}'.format(result[1]))
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis, reject the null hypothesis.Data is stationary")
    else:
        print("weak evidence against null hypothesis,indicating it is non-stationary ")
```

```python
adf_test(ts)
```

#### Making the series stationery

#### Train - Test split

```python
size = int(len(ts) * 0.8)
```

```python
ts_train = ts.iloc[:size]
```

```python
ts_test = ts.iloc[size:]
```

```python
ts_train.tail()
```

```python
ts_test.head()
```

#### Building ARIMA model

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
model = ARIMA(ts_train,order = (1,1,1)).fit()
```

#### prediction

```python
len(ts_train)
```

```python
pred = model.predict(start = len(ts_train),end = (len(ts)-1))
```

```python
pred.head()
```

#### Model evaluation

```python
from sklearn.metrics import mean_squared_error
```

```python
err = np.sqrt(mean_squared_error(ts_test,pred))
err
```

```python
ts_train.plot(label = 'train',legend = True,figsize = (20,6))
ts_test.plot(label = 'test',legend = True)
pred.plot(label = 'ARIMA',legend = True)
```
